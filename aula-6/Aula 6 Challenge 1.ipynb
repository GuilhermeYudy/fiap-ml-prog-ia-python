Crie um novo dataframe correlacionando o id das tabelas csv “pimas-data-norm-2.Csv” com “pimas-data-city.Csv” na pasta data (Github).

```python
import numpy as np
import pandas as pd
```

```python
df_pimas = pd.read_csv("data/pima-data-norm-2.csv", index_col=0)
df_pimas.head(5)
```

```python
df_pimas_city = pd.read_csv("data/pima-data-city.csv", sep=";", index_col=0)
df_pimas_city.head(5)
```

```python
df_pimas_data_city = pd.merge(df_pimas, df_pimas_city, left_index=True, right_index=True)
df_pimas_data_city.head(5)
```

```python
df_pimas_data_city.to_csv("data/pima-data-norm-city.csv")
```
